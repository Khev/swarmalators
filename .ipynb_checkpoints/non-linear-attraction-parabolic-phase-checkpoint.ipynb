# Introduction

Janus particles form a beating cluster state in which a cluster of synced particles expand and contracts periodically (see papers directory). Can I recover this in the swarmalator model? What effects do you need?

1. Distributed omega
2. Sigma small
3. J > 1 --> phase dependent repulsion (this is most likely; see the paper)

I tried the linear parabolic model, but the unphysical behavior as r -> infintity produced absurdities. 

```python
import time
import numpy as np
import swarmalators_funcs as f
from scipy.integrate import odeint
import matplotlib as mpl
from matplotlib  import cm
import matplotlib.pyplot as plt
from IPython.display import clear_output, display, HTML 
from IPython.html.widgets import interact, interactive 
%matplotlib inline

def rhs_unit_vector(z,t,J,K,sigma,n,omega):
    """
    z0 = [x0, y0, theta0], where x0[i], y0[i], theta0[i] gives the x[t=0], y[y=0], theta[t=0] of 
    the i-th swarmalator
    t = time parameter (need this here to pass to python's odeint function)
    J = float
    K = float
    n = number of swarmalators
    omega = [omega1, omega2, ..., ] natural frequencies of swarmaltors
    """
    
    #Instantiate -- set up 
    x = z[0:n]
    y = z[n:2*n]
    theta = z[2*n:3*n]
    
    # Set up as a numpy.array to make the computation faster
    xd = x[:, np.newaxis] - x
    yd = y[:, np.newaxis] - y
    theta_d = theta[:, np.newaxis] - theta
    dist_sq = xd**2 + yd**2
    inverse_dist_sq = np.nan_to_num(1.0/dist_sq)
    inverse_dist_4 = np.nan_to_num(1.0/dist_sq**2)
    np.fill_diagonal(inverse_dist_sq, 0.0)  ## correct 1 / d_ii = 1 / o error
    parabolic_dist = (1 - dist_sq / (sigma))
    parabolic_dist = parabolic_dist * np.heaviside(sigma*np.ones_like(dist_sq) - dist_sq, dist_sq)
    
    x_rhs = -1*np.nan_to_num( xd*( (1+J*np.cos(theta_d))*inverse_dist_sq - inverse_dist_4)) 
    y_rhs = -1*np.nan_to_num( yd*( (1+J*np.cos(theta_d))*inverse_dist_sq - inverse_dist_4)) 
    theta_rhs = -1*K*np.sin(theta_d)*parabolic_dist
    
    x_next = np.nan_to_num((1/float(n))*np.sum((1-np.eye(xd.shape[0]))*x_rhs, axis=1))
    y_next = np.nan_to_num((1/float(n))*np.sum((1-np.eye(xd.shape[0]))*y_rhs, axis=1))
    theta_next = np.nan_to_num(omega +  (1/float(n))*np.sum((1-np.eye(xd.shape[0]))*theta_rhs, axis=1))
                                              
    return np.concatenate((x_next, y_next, theta_next))
```

### Sim

- I dont think unit vector is going to cut it
- Need to keep sigma small
- I need a phase gradient across the disk, so that the core is repulsed from the rim. For this I may need to change (1 + cos(theta)) to something narrower
- It could be the J < 1, this requires a strong phase gradientin the r direction. If there are assymeties, this can destabilize the state. So I think its more likely its a delay term. The heterogeneity in omega break the symmetry

```python
# Simulation parameters
np.random.seed(0)
a, dt, T, n, L  = 1, 0.5, 100, 30, 1   # surprisingly, dt = 0.5 seems to work OK (for prelimiart)
x0 = np.random.uniform(-L,L,n);y0=np.random.uniform(-L,L,n);theta0 = np.random.uniform(-np.pi,np.pi,n)
J, K = 2.0, 10
sigma, omega = 0.5, 0.2
#omega = np.array([omega for i in range(n)])
#omega = np.random.uniform(-omega,omega,n)
#omega = np.linspace(-omega,omega,n)
omega = np.random.normal(0,omega,n)
z0 = [x0, y0, theta0]
t = [dt*i for i in range(int(T/dt))]

# Do simulation
tic = time.clock()
z0 = np.array([x0,y0,theta0]);z0 = z0.flatten()
sols = odeint(rhs_unit_vector, z0, t, args=(J,K,sigma,n,omega))
x, y, theta = f.unpack(sols,n)

# Collect auxiliary info, like the order parameter gamma, and S
r, phi = f.cart_to_polar(x,y)
R, R1 = np.abs(f.find_Z(x,y,theta)), np.abs(f.find_Z1(x,y,theta))
Sp, Sm = np.abs(f.find_W_plus(x,y,theta)), np.abs(f.find_W_minus(x,y,theta))
a_max = [np.max( np.sqrt(x[t,:]**2 + y[t,:]**2) ) for t in range(x.shape[0])]
a_min = [np.min( np.sqrt(x[t,:]**2 + y[t,:]**2) ) for t in range(x.shape[0])]
toc = time.clock()
print('took {:.2f} min'.format((toc-tic)/60.0))

# Plot at end
plt.figure(figsize=(15,5))
plt.subplot(131)
norm, cmap = mpl.colors.Normalize(vmin=0, vmax=2*np.pi), cm.gist_rainbow
m = cm.ScalarMappable(norm=norm, cmap=cmap)
temp = m.to_rgba(np.mod(theta[-1,:],2*np.pi))
plt.scatter(x[-1,:],y[-1,:], c = temp, s = 200, alpha = 0.9, marker = 'o',edgecolors='none', 
            cmap = cm.gist_rainbow)

plt.subplot(132)
plt.plot(R);plt.plot(R1);plt.plot(Sp);plt.plot(Sm)
plt.legend(['R0', 'R1','S+','S-']);plt.ylim([0,1.1])
plt.subplot(133)
plt.plot(a_max);plt.plot(a_min);
plt.legend(['Radius max', 'Radius min']);plt.ylim([0,4])
```

### Interactive 

Can't use this if you've not loaded the last three module (the interactive one) in my preamble.

```python
def scatter_t(x,y,theta,t):
    """ Make a scatter plot of swarmalators at time t in the (x,y) plane 
        where swarmalators are colored according to their phase
        
        x = np.array, x[t][i] = x-coord of i-th swarmalator at time t
        y = np.array, y[t][i] = y-coord of i-th swarmalator at time t
        
    """
    
    fig = plt.figure(num=None, figsize=(10, 6), dpi=80, facecolor='w', edgecolor='k')
    norm = mpl.colors.Normalize(vmin=0, vmax=2*np.pi)
    cmap = cm.gist_rainbow
    m = cm.ScalarMappable(norm=norm, cmap=cmap)
    temp = m.to_rgba(np.mod(theta[t,:],2*np.pi))
    plt.scatter(x[t,:],y[t,:], c = temp, s = 200, alpha = 0.9, marker = 'o',edgecolors='none',cmap = cm.gist_rainbow)
    plt.xlim([-4,4]);plt.ylim([-4,4])
    
def plot_temp(t):
    scatter_t(x,y,theta,t)
    
interact(plot_temp,t=(0,x.shape[0]-1,1))
```
