# Introduction

Here I want to check if the linear parabolic model has the same states as the origianl model 

```python
import numpy as np
import swarmalators_funcs as f
from scipy.integrate import odeint
import time
import matplotlib.pyplot as plt
from IPython.display import clear_output, display, HTML 
from IPython.html.widgets import interact, interactive 
%matplotlib inline

def rhs_linear_spatial_parabolic_phase(z,t,J,K,sigma,n,omega):
    
    # Instantiate
    x = z[0:n]
    y = z[n:2*n]
    theta = z[2*n:3*n]
    xd = x[:, np.newaxis] - x
    yd = y[:, np.newaxis] - y
    theta_d = theta[:, np.newaxis] - theta
    inverse_dist_sq = np.nan_to_num(1.0/((xd)**2 + (yd)**2))
    np.fill_diagonal(inverse_dist_sq, 0.0) ## correct 1 / d_ii = 0 / 1 erro
    dist_sq = xd**2 + yd**2
    parabolic_dist = (1 - dist_sq / (sigma))
    parabolic_dist = parabolic_dist * np.heaviside(sigma*np.ones_like(dist_sq) - dist_sq, dist_sq)
    
    x_rhs = -1*np.nan_to_num( xd*( (1+J*np.cos(theta_d)) - inverse_dist_sq)  ) 
    y_rhs = -1*np.nan_to_num( yd*( (1+J*np.cos(theta_d)) - inverse_dist_sq)  ) 
    theta_rhs = -1*K*np.sin(theta_d)*parabolic_dist
    
    # RHS
    x_next = np.nan_to_num((1/float(n))*np.sum((1-np.eye(xd.shape[0]))*x_rhs, axis=1))
    y_next = np.nan_to_num((1/float(n))*np.sum((1-np.eye(xd.shape[0]))*y_rhs, axis=1))
    theta_next = np.nan_to_num(omega +  (1/float(n))*np.sum((1-np.eye(xd.shape[0]))*theta_rhs, axis=1))
                                              
    return np.concatenate((x_next, y_next, theta_next))
```

### Do simulation

```python
# Simulation parameters
a, dt, T, n, L  = 1, 0.5, 50, 200, 1   # surprisingly, dt = 0.5 seems to work OK (for prelimiart)
np.random.seed(0)
x0 = np.random.uniform(-L,L,n);y0=np.random.uniform(-L,L,n);theta0 = np.random.uniform(-np.pi,np.pi,n)
#theta0 = np.array([0 for i in range(n)])
omega, sigma = np.zeros(n), 2.0  # 2 sigma might be a more natural choice
J,K = 1.0, 0      #static phase wave 
z0 = [x0, y0, theta0]
t = [dt*i for i in range(int(T/dt))]

# Do simulation
tic = time.clock()
z0 = np.array([x0,y0,theta0])
z0 = z0.flatten()
sols = odeint(rhs_linear_spatial_parabolic_phase, z0, t, args=(J,K,sigma,n,omega))
x, y, theta = f.unpack(sols,n)

# Collect auxiliary info, like the order parameter gamma, and S
r, phi = f.cart_to_polar(x,y)
transient_index = int(0.5*x.shape[0])
gamma = f.find_gamma(phi[transient_index:])
v = f.find_vel_t(x,y,theta,dt)
v_mean = np.mean([np.mean(v[t,:]) for t in range(transient_index,v.shape[0])])
R = np.mean(np.abs(f.find_Z(x,y,theta))[transient_index:-1])
S_plus = np.mean(np.abs(f.find_W_plus(x,y,theta))[transient_index:-1])
S_minus = np.mean(np.abs(f.find_W_minus(x,y,theta))[transient_index:-1])
Rmin, Rmax = [min(r[t,:]) for t in range(r.shape[0])], [max(r[t,:]) for t in range(r.shape[0])]
Rmin, Rmax = np.mean(Rmin[transient_index:]), np.mean(Rmax[:transient_index])
toc = time.clock()
print('took {:.2f} min'.format((toc-tic)/60.0))
print('gamma, R, S+, S- = {:.2f}, {:.2f}, {:.2f}, {:.2f}'.format(gamma, R, S_plus, S_minus))
print('Rmin, Rmax = {:.2f}, {:.2f}'.format(Rmin,Rmax))

# Plot at end
f.scatter_t(x,y,theta,-1)
```

### Test mid radius

$ r_{mid} = \sqrt{ \frac{J \sigma}{-2K} } $

```python
# Simulation parameters
Ks = np.linspace(-5,0,40)
Rmids, Rmins, Rmaxes, Sps, Sms, gammas = [], [], [], [], [], []
for K in Ks:
    np.random.seed(0)
    a, dt, T, n, L  = 1, 0.5, 500, 100, 1   # surprisingly, dt = 0.5 seems to work OK (for prelimiart)
    x0 = np.random.uniform(-L,L,n);y0=np.random.uniform(-L,L,n);theta0 = np.random.uniform(-np.pi,np.pi,n)
    J, omega, sigma = 1.0, np.zeros(n), 2.0  # 2 sigma might be a more natural choice
    z0 = [x0, y0, theta0]
    t = [dt*i for i in range(int(T/dt))]

    # Do simulation
    tic = time.clock()
    z0 = np.array([x0,y0,theta0])
    z0 = z0.flatten()
    sols = odeint(rhs_linear_spatial_parabolic_phase, z0, t, args=(J,K,sigma,n,omega))
    x, y, theta = f.unpack(sols,n)

    # Collect auxiliary info, like the order parameter gamma, and S
    r, phi = f.cart_to_polar(x,y)
    transient_index = int(0.9*x.shape[0])
    gamma = f.find_gamma(phi[transient_index:])
    v = f.find_vel_t(x,y,theta,dt)
    v_mean = np.mean([np.mean(v[t,:]) for t in range(transient_index,v.shape[0])])
    R = np.mean(np.abs(f.find_Z(x,y,theta))[transient_index:-1])
    S_plus = np.mean(np.abs(f.find_W_plus(x,y,theta))[transient_index:-1])
    S_minus = np.mean(np.abs(f.find_W_minus(x,y,theta))[transient_index:-1])
    Rmin, Rmax = [min(r[t,:]) for t in range(r.shape[0])], [max(r[t,:]) for t in range(r.shape[0])]
    Rmin, Rmax = np.mean(Rmin[transient_index:]), np.mean(Rmax[:transient_index])
    Rmid = (Rmax+Rmin) / 2.0
    Rmids.append(Rmid)
    Rmaxes.append(Rmax)
    Rmins.append(Rmin)
    Sps.append(S_plus);Sms.append(S_minus)
    gammas.append(gamma)
    toc = time.clock()
    print('K = {:.2f} took {:.2f} min'.format(K,(toc-tic)/60.0))
```

```python
plt.figure(figsize=(10,8))
plt.subplot(311)
plt.plot(-Ks,Sps);plt.plot(-Ks,Sms);plt.plot(-Ks,gammas);
plt.vlines(1.2,0,2,linestyle='dashed');plt.ylim([0,1.1])
plt.xlabel('-K', fontsize=14);plt.legend(['S+','S-','gamma'])

plt.subplot(312)
plt.plot(-Ks,Rmaxes);plt.plot(-Ks,Rmins);
plt.vlines(1.2,0,2,linestyle='dashed');
plt.xlabel('-K', fontsize=14);plt.legend(['Rmax','Rmin'])

plt.subplot(313)
plt.plot(-Ks,[np.sqrt(-(J*sigma)/(2*K)) for K in Ks])
plt.plot(-Ks,Rmids,'o')
plt.xlabel('-K', fontsize=14);plt.ylabel('Rmid', fontsize=14)
plt.legend(['theory','data'])
```

```python
# r_mid^2 = J*sigma / (-2*K)
plt.plot(-Ks,[np.sqrt(-(J*sigma)/(2*K)) for K in Ks])
plt.plot(-Ks,Rmids,'o')
plt.xlabel('-K', fontsize=14);plt.ylabel('Rmid', fontsize=14)
plt.legend(['theory','data'])
plt.xlim([1.5,3]);plt.ylim([0,1])
```

### Test rmin, rmax

$ H(r^*) = \frac{J S}{2}  $ 


where 

$ H(r,\phi) = \frac{1}{N} \sum_j (r_j \cos(\phi_j - \phi_i) - r_i) ( 1 - 1 / d_{ij}^2  ) $

```python
t = -10
Hs, rs, phis = [], [], []
for i in range(n):
    H = 0
    for j in range(n):
        if j != i:
            ri, rj = r[t,i], r[t,j]
            phii, phij = phi[t,i], phi[t,j]
            term1 = rj*np.cos(theta[t,j]-theta[t,i]) - ri
            term2 = 1.0 - 1.0 / ( ri**2 -2*ri*rj*np.cos(phii-phij) + rj**2 )
            H += term1*term2
    H /= (1.*n)
    Hs.append(H)
    rs.append(ri)
    phis.append(phii)
plt.figure(figsize=(15,5))
plt.subplot(121)
plt.plot(rs,Hs,'.')
plt.subplot(122)
plt.plot(phis,Hs,'.')
```

```python
fig = plt.figure(figsize=(9,6))
ax = fig.add_subplot(111, projection='polar')
_ = ax.scatter(phis, rs, c=Hs, cmap='hsv', alpha=0.75)
#fig.colorbar(Hs, ax=ax)
```
