# Introduction

Here I want to check if the linear parabolic model has the same states as the origianl model 

```python
import numpy as np
import swarmalators_funcs as f
from scipy.integrate import odeint
import time
import matplotlib.pyplot as plt
from IPython.display import clear_output, display, HTML 
from IPython.html.widgets import interact, interactive 
%matplotlib inline

def rhs_linear_spatial_parabolic_phase(z,t,J,K,sigma,n,omega):
    
    # Instantiate
    x = z[0:n]
    y = z[n:2*n]
    theta = z[2*n:3*n]
    xd = x[:, np.newaxis] - x
    yd = y[:, np.newaxis] - y
    theta_d = theta[:, np.newaxis] - theta
    inverse_dist_sq = np.nan_to_num(1.0/((xd)**2 + (yd)**2))
    np.fill_diagonal(inverse_dist_sq, 0.0) ## correct 1 / d_ii = 0 / 1 erro
    dist_sq = xd**2 + yd**2
    parabolic_dist = (1 - dist_sq / (sigma))
    parabolic_dist = parabolic_dist * np.heaviside(sigma*np.ones_like(dist_sq) - dist_sq, dist_sq)
    
    x_rhs = -1*np.nan_to_num( xd*( (1+J*np.cos(theta_d)) - inverse_dist_sq)  ) 
    y_rhs = -1*np.nan_to_num( yd*( (1+J*np.cos(theta_d)) - inverse_dist_sq)  ) 
    theta_rhs = -1*K*np.sin(theta_d)*parabolic_dist
    
    # RHS
    x_next = np.nan_to_num((1/float(n))*np.sum((1-np.eye(xd.shape[0]))*x_rhs, axis=1))
    y_next = np.nan_to_num((1/float(n))*np.sum((1-np.eye(xd.shape[0]))*y_rhs, axis=1))
    theta_next = np.nan_to_num(omega +  (1/float(n))*np.sum((1-np.eye(xd.shape[0]))*theta_rhs, axis=1))
                                              
    return np.concatenate((x_next, y_next, theta_next))
```

### Do simulation

```python
# Simulation parameters
a, dt, T, n, L  = 1, 0.5, 500, 200, 1   # surprisingly, dt = 0.5 seems to work OK (for prelimiart)
np.random.seed(0)
x0 = np.random.uniform(-L,L,n);y0=np.random.uniform(-L,L,n);theta0 = np.random.uniform(-np.pi,np.pi,n)
#theta0 = np.array([0 for i in range(n)])
omega, sigma = np.zeros(n), 5.0  # 2 sigma might be a more natural choice
#J,K = 0.1, 1      #static sync state
#J,K = 0.1, -1     #static async state
#J,K = 1.0, 0       #static phase wave 
#J,K = 1.0, -0.1,  #splintered phase wave --- note, long transient here
#J,K = 1.0, -0.75  #active phase wave 
J,K = 1.0, -50.0      #static phase wave 
z0 = [x0, y0, theta0]
t = [dt*i for i in range(int(T/dt))]


# Do simulation
tic = time.clock()
z0 = np.array([x0,y0,theta0])
z0 = z0.flatten()
sols = odeint(rhs_linear_spatial_parabolic_phase, z0, t, args=(J,K,sigma,n,omega))
x, y, theta = f.unpack(sols,n)

# Collect auxiliary info, like the order parameter gamma, and S
r, phi = f.cart_to_polar(x,y)
transient_index = int(0.9*x.shape[0])
gamma = f.find_gamma(phi[transient_index:])
v = f.find_vel_t(x,y,theta,dt)
v_mean = np.mean([np.mean(v[t,:]) for t in range(transient_index,v.shape[0])])
R = np.mean(np.abs(f.find_Z(x,y,theta))[transient_index:-1])
S_plus = np.mean(np.abs(f.find_W_plus(x,y,theta))[transient_index:-1])
S_minus = np.mean(np.abs(f.find_W_minus(x,y,theta))[transient_index:-1])
toc = time.clock()
print('took {:.2f} min'.format((toc-tic)/60.0))
print('gamma, R, S+, S- = {:.2f}, {:.2f}, {:.2f}, {:.2f}'.format(gamma, R, S_plus, S_minus))

# Plot at end
f.scatter_t(x,y,theta,-1)
```

### Sweep over K in the state

```python
a, dt, T, n, L  = 1, 0.5, 500, 100, 1 
t = [dt*i for i in range(int(T/dt))]
np.random.seed(0)
x0 = np.random.uniform(-L,L,n);y0=np.random.uniform(-L,L,n);theta0 = np.random.uniform(-np.pi,np.pi,n)
J, sigma, omega = 1.0, 2.0, np.zeros(n);

outs = []
Ks = np.linspace(-10,-5,10)
for K in Ks:
    tic = time.clock()
    z0 = np.array([x0,y0,theta0])
    z0 = z0.flatten()
    sols = odeint(rhs_linear_spatial_parabolic_phase, z0, t, args=(J,K,sigma,n,omega))
    x, y, theta = f.unpack(sols,n)
    r, phi = f.cart_to_polar(x,y)
    gamma = f.find_gamma(phi)
    v = f.find_vel_t(x,y,theta,dt)
    transient_index = int(0.9*x.shape[0])
    v_mean = np.mean([np.mean(v[t,:]) for t in range(transient_index,v.shape[0])])
    R0 = np.mean(np.abs(f.find_Z(x,y,theta))[transient_index:-1])
    R1 = np.mean(np.abs(f.find_Z1(x,y,theta))[transient_index:-1])
    S_plus = np.mean(np.abs(f.find_W_plus(x,y,theta))[transient_index:-1])
    S_minus = np.mean(np.abs(f.find_W_minus(x,y,theta))[transient_index:-1])
    #R_max = np.mean([np.max( np.sqrt(x[t,:]**2 + y[t,:]**2) ) for t in range(x.shape[0])][transient_index:-1])
    #R_min = np.mean([np.min( np.sqrt(x[t,:]**2 + y[t,:]**2) ) for t in range(x.shape[0])][transient_index:-1])
    out = [gamma, v_mean, S_plus, S_minus, R0, R1]
    outs.append(out)
    print('K = {:.2f} done'.format(K))
outs = np.array(outs)

# Plot
plt.figure(figsize=(15,5))
plt.subplot(231)
plt.plot(Ks,outs[:,0],'o-')
plt.ylabel('gamma')

plt.subplot(232)
plt.plot(Ks,outs[:,2],'o-')
plt.ylabel('S+')
plt.ylim([0,1])
plt.title('(J, sigma, omega) = ({},{},{})'.format(J,sigma,omega[0]))

plt.subplot(233)
plt.plot(Ks,outs[:,3],'o-')
plt.ylabel('S-')
plt.ylim([0,1])

plt.subplot(234)
plt.plot(Ks,outs[:,1],'o-')
plt.xlabel('-K')
plt.ylabel('<v>')

plt.subplot(235)
plt.plot(Ks,outs[:,-2],'o-')
plt.xlabel('-K')
plt.ylabel('R0')
plt.ylim([0,1])

plt.subplot(236)
plt.plot(Ks,outs[:,-1],'o-')
plt.xlabel('-K')
plt.ylim([0,1])
_ = plt.ylabel('R1')
```

 - So it looks like the order parameter never drops to zero
 - So there isn't a transition. This could change as I increase N
 - I'll put N = 500, and see does S_- decrease
